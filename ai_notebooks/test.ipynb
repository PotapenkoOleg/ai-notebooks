```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%config InlineBackend.figure_format = 'retina'
```

```python
data = {'size': pd.Series([1700, 2100, 1900, 1300, 1600, 2200]),
        'price': pd.Series([51000, 63000, 57000, 39000, 48000, 68000])}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(data=df, x='size', y='price');
```

```python
data = {'size': pd.Series([1700, 2100, 1900, 1300, 1600, 2200]),
        'price': pd.Series([53000, 65000, 59000, 41000, 50000, 68000])}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(data=df, x='size', y='price');
```

```python
sns.regplot(data=df, x='size', y='price', fit_reg=True, line_kws={'color': 'red'});
```
