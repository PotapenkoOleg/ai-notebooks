```python
import numpy as np
import matplotlib.pyplot as plt

# Wavelengths of visible light in nanometers
wavelengths = np.linspace(380, 750, 1000) # from violet to red

# Rayleigh scattering proportionality: I ~ (1/^4)
intensity_scattering = 1 / wavelengths**4

# Normalize the intensity for better visualization
intensity_scattering_normalized = intensity_scattering / np.max(intensity_scattering)

plt.figure(figsize=(10, 6))
plt.plot(wavelengths, intensity_scattering_normalized, label="Scattering Intensity")
plt.axvline(x=475, color='blue', linestyle='--', label='Approx. Blue Light (~475 nm)')
plt.title("Rayleigh Scattering in the Atmosphere")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Normalized Scattering Intensity")
plt.legend()
plt.grid(True)
plt.show()
```
